```python
counts, bin_edges = np.histogram(liv['nodes'], bins=10, density = True)

pdf = counts/(sum(counts))

print(pdf)

print(bin_edges)

cdf = np.cumsum(pdf)

plt.plot(bin_edges[1:],pdf)

plt.plot(bin_edges[1:], cdf)
```
